# ラベルを付けて、学習に使うデータセットを作成する

```python
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import seaborn as sns
import pandas as pd
import sys
import stockpy as st
from datetime import datetime
from tqdm import tqdm
```

```python
df = pd.read_csv('./train_dataset_by_time2.csv', header=0)
```

```python
def process_data(data):
    try:
        data.pop()
        if ((0.02 < data[-1] - data[-2]) and (0.02 < data[-2] - data[-3])):        
            data.insert(0, 1)
        else:
            data.insert(0, 0)
        # リストをNumPy配列に変換
        data = np.array(data)
        return data
    except Exception as e:
        print("Error processing data")
        print(e)
        raise Exception
```

```python
df3 = pd.DataFrame()
rows = []
for row in tqdm(df.itertuples(index=False)):
    row = process_data(list(row))
    rows.append(row)

df3 = pd.DataFrame(rows)
```

```python
df3.to_csv('labeled_train_dataset2.csv', index=False)
```

```python
# カテゴリごとのカウントを取得
value_counts = df3[0].value_counts()
# 最も少ないカウント数を取得
min_count = value_counts.min()
# 最も少ないカウント数のカテゴリを取得
min_count_categories = value_counts[value_counts == min_count].index
# フィルタリング
filtered_df_1 = df3[df3[0].isin(min_count_categories)]
# カテゴリごとのカウントを取得
value_counts = df3[0].value_counts()
# 最も少ないカウント数を取得
max_count = value_counts.max()
# 最も少ないカウント数のカテゴリを取得
max_count_categories = value_counts[value_counts == max_count].index
# フィルタリング
filtered_df_2 = df3[df3[0].isin(max_count_categories)][:min_count]
result = pd.concat([filtered_df_1, filtered_df_2], axis=0)
```
